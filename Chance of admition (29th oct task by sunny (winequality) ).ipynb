## Let's Slicing the dependent and independent variables

```python
x = data.iloc[:,:8]
y = data.iloc[:,-1]
```

let's split the data into train and test data.

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size=.33, random_state=42)
```

# First standardize the data using Standard scaler:

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
```

```python

xtrain_fit_transform = scale.fit_transform(xtrain)
xtest_transform = scale.transform(xtest)
```

```python
xtrain_fit_transform
```

```python
xtest_transform
```

```python
ytrain
```

```python
ytest
```

# Build the model

```python
from sklearn.svm import SVR
svr_model = SVR()
```

```python
svr_model.fit(xtrain_fit_transform, ytrain)
```

```python
predictions = svr_model.predict(xtest_transform)
predictions
```

```python
svr_model.score(xtrain_fit_transform,ytrain)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, r2_score
```

```python
ytest.shape
```

```python
predictions.shape
```

Dependent variable is continuos, so we have to use r2 score.
for classification problems we hace to use accuracy_score.

```python
score = r2_score(ytest, predictions)
score
```
